#### 1. 计算机基础

##### 1.1 计算机硬件的世界

**中央处理器（CPU）：大脑指挥官**  
CPU可以理解为计算机的大脑。一切决策、计算、如何安排工作流程、指挥其他硬件，都由它完成。就像艺术创作时的主脑，有思路、有想法、能“算计”，并且能下命令。

**内存（RAM）：短期记忆管家**  
内存类似于我们的“短期记忆”——比如演奏一段乐曲时记住下一个音符，画素描时记得刚刚涂过的那一笔。它把近期需要频繁访问的信息临时存放，关机一刹那它就全部“清零”。你可以想象成是一块巨大的白板，随时可以擦去重写。

**硬盘/固态硬盘：作品库与日记**  
硬盘就像自己的工作室或画作档案柜。里面储存着你的所有成品、草图、计划，即使很久没开电脑，内容也会一直存在。

**输入设备和输出设备：感官与表达**  
键盘、鼠标、画板是计算机“感知外界”的方式，类似你的手眼耳。显示器、音箱、打印机则是把内部想法、结果展现给世界的“嘴巴”和“表情”。

**冯诺依曼结构**
冯诺依曼结构是现代计算机的经典设计方案，可以把它想象成一套高效的“工作流程”。  
在这套结构中，计算机的全部任务都是靠五大核心部件协作完成的：

- **输入设备**：负责把数据送进计算机，比如键盘、鼠标，就像递交原材料的助手。
- **输出设备**：负责把计算结果展示出来，如显示器、打印机，就像把成品展现给大家。
- **存储器**：是计算机用来记东西的地方，既能存数据，也能存程序，相当于资料库和工作计划本。
- **运算器**：用来进行各种计算和判断，相当于动手干活的助手。
- **控制器**：负责协调和指挥各个部件的工作，像工作室里的导演或老师。

**核心思想**：所有的数据和程序都统一存储在存储器中，计算机按顺序去执行每一个步骤。  
这一结构让计算机变得灵活、通用，能胜任复杂多样的任务。


##### 1.2 内存与地址

**内存空间的想象**  
可以把内存想象成一排房间，每个房间只存专门的数据，比如一管颜料、一张便签。每个房间的“门牌号”就是它的**地址**。

**地址唯一性**  
房间号码（地址）都不重复，所以只要你记得门牌号，随时回去就能找到那一份数据。之后学习“指针”时，指针其实就是帮你记住地址、带路的“钥匙”或“导航卡”，指到具体的内存房间。

**数据的存放与查找**  
像“x = 10”这种变量，本质上是在某个房间存放了数字10。访问x？就是拿着钥匙去那间房看看。

**地址线/总线如何影响“房间”数量**  
物理上每加一根“地址线”，“房间”数量就翻倍。如果有10根线，最多能有1024间房。如果是20根，居然可以有一百万多间！


##### 1.3 进制系统

**为何需要进制？**  
我们的世界基本用十进制（每十进一），但计算机的物理结构只懂0和1（有/无），所以天生“母语”是二进制。有时候，为了表达色彩、像素等内容，又用十六进制，既简洁又直观。

**最基础的进制种类**  
- **二进制**：只有0和1。  
- **十进制**：0到9。  
- **十六进制**：0–9和A–F（多见于色彩编码，例如#FF00AA）。
- 生活小比喻：就像一支舞蹈用不同的动作元素来“拼词造句”；进制是数字世界的“语法”。

**二进制的实际意义**  
每一个0或1称为“一位”（bit）。8位就是一个字节（byte）。一个字节最多可以表示256种情况。二进制描述一切数字，在底层没有“中间态”。

**十六进制与艺术设计**  
常见于颜色表示：“#FF66CC”其实就是三组二进制数字组成的（R、G、B值），简洁高效，很适合程序和设计师对话。

#### 2. 终端操作

无论你是刚接触计算机的艺术生，还是未来准备成为技术大拿的开发者，都必须清楚：**“会用终端和命令行”是通向高效编程和掌控计算机世界的必经之路。**

你是否有这样的体验：  
- 看到那一堆“字母构成”的窗口，觉得如入无人之境，自己好像离“会编程”还差很远？
- 用图形界面点来点去，一旦出问题就无计可施？
- 在网上找教程时，总发现很多关键步骤都让你“复制一行代码到你的终端”？

原因就是，**终端和命令行**才是计算机深度交互的“原生渠道”，它给了你对文件、程序、系统的完整掌控。图形界面只是“包装纸”，而终端才是“发动机”。    

想象一下，那些上世纪的科幻电影中（对，就是星球大战这种），航天员是如何与系统交流的，再想想现代的电影中，乘客又是如何与人工智能交流的。人机交互的发展史大致可以分为以下几个阶段：

- **打孔卡片时代**：“话语”只能藏在一张张镂空的纸片里，计算机读卡片，不会“听”也不会“说”。  
- **命令行终端诞生**：电脑终于有了“听见”我们输入指令的能力。我们用键盘敲出一行又一行的命令，每个字母，每个符号，都是直接给计算机下达的命令。
- **图形界面亮相**：鼠标一动，点一点——不需要记忆复杂指令，和电脑的对话变得“看得见、摸得着”。
- **触控、语音交互**：今天，只需轻点屏幕、甚至直接说句话，计算机就能明白你的意图。

终端的原义是，计算机的“终端设备”，负责把信息显示给人类，并把人的输入传给主机。
而随着计算机硬件技术、与软件界面（UI）技术的发展，现代终端的含义变得更加丰富：
- 物理终端：比如显示器、手机、打印机等末端设备
- 终端仿真器（Terminal Emulator）：iTerm2、Windows Terminal、GNOME Terminal 等软件，用软件模拟老式硬件终端的通信协议（最常见是 ANSI/VT100）。

在现代计算机系统中，终端（Terminal） ≠ Shell ≠ 命令行（CLI）
• 终端：输入输出窗口
• Shell：解释你输入的文本并调用内核的程序（bash、zsh、fish、PowerShell…）
• 命令行：Shell 内部的语言和可执行程序

命令行，其实就像用很简练的外语在和计算机交流，句式结构也有规律可循：

```
命令   [选项]   [参数]
动词   “怎么做”  “对谁做”
```
比如：

- `ls -l Documents`  
列出文件 -同时列出详细信息 Document文件夹里的
- `cp file1.txt file2.txt`  
复制 file1.txt 为 file2.txt  
- `mkdir new_home`  
创建文件夹 名称为new_home

常用命令行清单：

| 命令         | 作用                          |
| ---------- | --------------------------- |
| pwd        | 显示当前所在的目录（即“你的人在文件夹树的哪个地方”）|
| ls         | 列出当前目录下的内容            |
| cd         | 改变当前目录（走到别的文件夹里）      |
| mkdir      | 创建新文件夹                    |
| touch      | 新建一个空文件或更新文件时间戳         |
| cp         | 复制文件                        |
| mv         | 移动文件或重命名                  |
| rm         | 删除文件（**谨慎使用**）             |
| cat        | 显示文件内容                      |
| man        | 查看命令的“使用说明书”              |

#### 3. 网络基础

最初，计算机像一座座孤岛，信息无法互通。随着发展，人们发现：**如果它们能交换数据，就像人们邮寄信件、打电话一样，世界会变得更高效和有趣。**  
于是，我们造出连接多台计算机的“网络”——让信息能在不同机器之间进行传递。

- 单机时代：数据交换靠软盘、U 盘，“搬运工”是你的双手。
- 串一根线：两台电脑用串口/并口互连，0-1 可以直接流动。
- 多台电脑：线越拉越乱，需要统一的“交通规则”——网络协议栈就诞生了。

最简版本的协议栈可分五层，每层各司其职，整体高效协作。

##### 物理层：信号的世界  
**作用**：负责怎样把“0和1”传送出去，比如通过电流、光、无线波。  
**最典型设备**：网线、光纤、集线器（Hub）。  
**形象比喻**：马路，保证快递车能顺利跑起来。

##### 数据链路层：本地身份和直接交友  
**作用**：让在同一个网络内的设备能互相“喊名字”并可靠地交换数据。这里的“名字”叫**MAC地址**——每台设备出厂时自带的独一无二的硬件编号。  
**最典型设备**：交换机（Switch）。  
**形象比喻**：在小区里，快递员凭快递单直送到家门口。

##### 网络层：世界地图与导航  
**作用**：设计全球范围的“地址和路线”，让数据可以跨越无数网络找到目标。**IP地址**就像全球快递的收件人准确地理标识。  
**最典型设备**：路由器（Router）。  
**形象比喻**：快递公司配送中心，决定包裹怎么中转、换路线、走哪条高速公路最快。

##### 传输层：数据的拆装和可靠性保障  
**作用**：保证从某台电脑的某个程序到另一台电脑的指定程序传送数据，能确保数据完整、排序正确。主要有两种方式——  
  - **TCP**：保证每条信息都送达，不丢不重。
  - **UDP**：快，但可能有遗漏和顺序错乱。  
**最典型设备**：这层主要由电脑里的操作系统负责；没有专门硬件。

##### 应用层：为用户打造的“沟通法则”  
**作用**：面向我们用的各种服务，比如看网页、收邮件、视频通话。这一层有不同的“沟通规则”，叫做“协议”，比如网站用HTTP、邮箱用SMTP等。  
**最典型设备**：各种应用服务器（网站服务器、邮箱服务器等）。

##### 总结

| 层级         | 关键词     | 解释                       | 典型设备         |
| ------------ | ---------- | -------------------------- | --------------- |
| 物理层       | 信号、电缆   | 传递 0 和 1 的通道           | 网线、Hub       |
| 数据链路层   | MAC地址    | 网络硬件的唯一编号           | 交换机Switch    |
| 网络层       | IP地址      | 全球唯一的“收发地址”         | 路由器Router    |
| 传输层       | 端口、TCP/UDP | 区分不同程序、保障可靠传输     | 操作系统功能    |
| 应用层       | 协议        | 用户实际操作对应的规则         | 各类服务器      |

计算机网络技术中，有一些十分常用的概念，这些内容也会出现在后续的网络编程技术中，这里对他们进行一个简单总结。

| 概念 | 所在层(OSI) | 功能 | 关键要点 |  
|------|-------------|------------|--------------------|  
| MAC 地址 | 数据链路层 | 局域网内唯一识别一块网卡 | “硬件身份证” |  
| IP 地址 | 网络层 | 把数据包送到哪台主机 | IPv4：32 位 |  
| 端口 | 传输层 | 在同一台主机内分配给进程 | 范围 0–65535 |  

| 协议 | 所在层(OSI) | 主要用途 | 关键特点 |  
|------|-------------|----------|----------------|  
| DHCP | 应用层 | 自动分配 IP、网关、DNS 等 | Client ➜ Discover → Server ➜ Offer → Client ➜ Request → Server ➜ Ack |  
| DNS | 应用层 | 把域名解析成 IP | 分层树状数据库；缓存优先，递归查询兜底 |  
| HTTP / HTTPS | 应用层 | Web 请求 / 响应 | 无状态；常见方法 GET/POST/PUT/DELETE |  

掌握 MAC、IP、端口、Socket 这些“地址与门牌号”后，就能在后续 Socket 编程中心中有图，调试时也能一眼看出数据卡在哪层。
这里有个一个口诀，帮助你记忆并区分不同的网络层内容：

**“网线（物理）接网卡（链路）→ 路由器（网络）→ 端口（传输）→ 浏览器/服务器（应用）”**

#### 4. 远程连接

前一节提到的终端，原本是人和计算机“对话”的窗口。最早它就是一块能显示字符的小黑屏，通过它我们输入命令，让计算机帮我们执行各种操作。在现代系统中，这一传统多演化成“命令行终端”，依然是开发和操作系统管理者的必备工具，尽管图形界面已随处可见，但终端的高效与精细还无法被替代。

随着网络普及，计算机早已不再是“单兵作战”。我们会遇到下面这些问题：

- 很多机器部署在其他房间、机房，或者根本不在同一栋楼里；
- 越来越多服务器都运行在云端数据中心，物理上接触不到；
- 有些设备，比如树莓派或公司服务器，长期无人值守，却偶尔要调试维护；
- 团队已远程协作办公，设备分布在世界各地。

此时，“远程接入”就变得必不可少——让你在自己电脑前，就能像坐在目标机器前一样操作和管理它们。

最常见的远程技术应用场景有：

- **远程桌面办公和支持**：公司IT帮你解决电脑难题，无需到场。
- **远程服务器维护**：开发者和运行维护人员能无障碍管理全球服务器。
- **资源共享/协作**：团队成员可同时访问同一台计算机资源。
- **物联网设备管理**：成百上千的智能终端分布各地，远程配置、高频监控，节省大量人力。

假如你家中有智能灯、空调和摄像头。你在外地旅行，发现家里可能没关空调，只需要点开手机App，远程连入家中的智能网关，随手“关掉”空调。这一过程中，你和冰箱、空调背后的“小计算机”正通过互联网实现了远程交互。

当下常见的远程接入方式有：

1. **SSH**（Secure Shell）：用于命令行远程管理，数据加密，平台通用。
2. **VNC**（Virtual Network Computing）：用于观看和操作远端的完整图形桌面。
3. **RDP**（Remote Desktop Protocol）：Windows自带的图形远程桌面协议。
4. **Web远程控制**：通过网页或专用客户端实现（如TeamViewer、Chrome 远程桌面等）。

这里详细介绍一下SSH与VNC技术。

##### SSH（安全外壳协议）

- **用途**：远程安全地登录另一台电脑、传送文件、运行指令。广泛用于Linux/Unix系统服务器。
- **原理**：全部数据均加密传输，防止中间窃听，支持密钥/口令等多种身份认证方式。
- **应用**：代码上传部署、远程重启、软件安装、系统维护等。
- **基本用法**：在终端输入  
  `ssh 用户名@目标主机IP`  
  例如：  
  `ssh alice@192.168.0.20`
- **扩展**：SCP/SFTP用于加密传文件，端口转发可做穿透或VPN。

##### VNC（虚拟网络计算）

- **用途**：远程操作完整的桌面环境，适合需要用到图形界面、鼠标键盘的场景（如给远程同事“演示”操作，或帮家人修电脑）。
- **原理**：远端计算机把桌面画面“拍照”发送到你这边，你的鼠标、键盘输入再被传回去，支持跨操作系统。
- **安全性**：大部分VNC协议本身不加密数据，常搭配SSH隧道提高安全（否则建议只在可信局域网内用）。
- **应用**：远程远程协助教学、图形化后台管理、不便接触物理终端时的界面控制。
- **跨平台**：Windows、Mac、Linux间都能互用。

#### 5. 学习资源

在学习编程和交互艺术的道路上，选择对了“伙伴”和“工具”，会让你的成长事半功倍。如今，越来越多优秀的学习平台和开发社区已经成为每一位计算机学习者不可或缺的助力：它们像不同的老师，有的手把手教基础，有的带你系统学习，有的则在你遇到困难时为你答疑解惑，有的则能让你亲身参与合作项目、见识真实的开发与创作过程。本节会为你介绍几类常用且强大的学习资源网站——它们可能会成为你未来几年甚至职业生涯的“常客”。希望你带着好奇和愿望，学会选用、用好这些平台，与全球各地的学习者和创作者同行。

- 菜鸟教程帮你快速起步；
- MOOC 让你系统学习；
- GitHub / Gitee 承载项目与协作；
- Stack Overflow 解决疑难杂症；
- LeetCode 刷题巩固算法。

##### 5.1 菜鸟教程

**菜鸟教程（runoob.com）** 是中文互联网最知名的程序设计速查与入门平台，面向初学者，涵盖了几乎所有主流编程语言、web开发、数据库以及各类常用技术。

- **亮点**：  
  - 全中文，免费
  - 简单明了，适合零基础入门
  - 代码示例丰富，可在线运行
- **适用场景**：快速查找基本语法、结构、常用函数，初学/忘记可以随时“现查现学”

##### 5.2 中国大学MOOC

**中国大学MOOC（中国慕课平台）**，网址 [icourse163.org](https://www.icourse163.org/)，是国内最大的视频公开课和自学门户。里面有大量来自全国一流高校和名师的计算机、交互艺术等课程。

- **特色**：  
  - 结构化完整、循序渐进
  - 有课件、作业和考试，适合“系统学习”或“自学力强化”
  - 有些课程会有助教在线答疑
- **适用场景**：打牢基础、课后自学、选修兴趣拓展、备考等

##### 5.3 Github

**Github** 是面向全球开发者的代码托管与协作平台。它不仅仅是“大众上传代码的网盘”，更是互联网上最重要的开源开发社区，也是学习范例、团队合作与技术交流的主阵地。

- **核心特色**：
  - 跨平台、跨国界代码协作
  - 每个项目可有历史记录、讨论区和问题追踪
  - 数百万开源项目可供参考和学习
- **核心功能**
  - 代码托管：通过 Git 管理版本；
  - Issues / Discussions：追踪 Bug 与需求；
  - Pull Request：代码评审与合并；
  - Actions：持续集成 / 自动部署。
- **初学者常用法**：
  - 搜索关键词，找到并阅读热门项目、入门资源、范例代码
  - 学习别人的项目结构和开发流程
  - `Star` 喜欢的项目，跟踪进展

##### 5.4 版本控制与 Git 简介

**版本控制** 就是帮你“记录和管理每一次改动历史”的工具，让你能随时回到过去，避免误操作导致代码丢失。**Git** 是当前最流行的分布式版本控制系统。

- **Git的基本用法**：  
  - 本地初始化：`git init`  
  - 提交一次更改：`git add .` -> `git commit -m "描述"`
  - 分支协作和代码合并、安全回滚
  - 上传到Github：`git push`
- **优势**：
  - 自动保留所有版本历史
  - 支持多人协作，精细记录每个人的更改
  - 常见于团队项目、开源社区和个人学习整理

##### 5.5 Gitee（码云）

**Gitee** 是中国开发者社区主推的代码托管和开源平台，界面、体验与Github类似，但访问速度更快，并有很多本土特色。

- **优势**：  
  - 完美支持中文、国内手机号注册  
  - 很多高校、企业的开源项目都在此托管，方便国内容易访问  
  - 有丰富的团队项目和专题代码集
- **适用场景**：  
  - 团队合作、发布自己的项目
  - 下载、查阅国内项目资料

##### 5.6 Stackoverflow

**Stackoverflow** 是全球程序员最大的技术问答和社区讨论平台。

- **功能**：  
  - 用户可提问任意编程疑难、报错，全球高手为你解答
  - 搜索几乎所有编程语言、工具的常见“坑”
- **使用建议**：  
  - 善用英文关键词，几乎所有新手踩到的坑，社区都有人遇到过，答案详细且权威
  - 注册账号后提问建议详细描述报错信息和环境，容易获得高质量帮助

##### 5.7 LeetCode

**LeetCode** 是全球主流的在线算法和编程练习平台，尤其在提高算法能力和“面试刷题”领域有极高声誉和资源积累。

- **特色**：  
  - 汇集上千道经典编程题目，各种难度分级，适合从零基础到进阶
  - 支持多种语言代码直接在线运行和测试
  - 可以看精选题解，学习他人思路
- **适合**：  
  - 算法提升、自主锻炼编程思维
  - 期中期末备考、工作面试前的刷题训练

#### 本章总结

##### 本章知识点汇总

1. 计算机基础  
   • 硬件五大核心：输入设备、输出设备、存储器、运算器(ALU)、控制器（冯诺依曼结构）。  
   • CPU＝“大脑”；RAM＝“短期记忆”；硬盘＝“长期档案”。  
   • 内存与地址：地址唯一、地址线决定可寻址空间；指针本质是“地址的变量”。  
   • 进制系统：二进制（机器语言）、十进制（人类习惯）、十六进制（高效表示 8-bit 数据）。

2. 终端（Terminal）与命令行  
   • 三者区别：Terminal＝窗口，Shell＝解释器，CLI＝命令语言。  
   • 命令结构：命令 + 选项 + 参数。  
   • 常用命令：pwd、ls、cd、mkdir、cp、mv、rm、cat、man 等。  
   • 掌握终端的意义：高效批量操作、自动化脚本、远程管理的基石。

3. 网络基础  
   • 五层模型速记：物理层(信号) → 数据链路层(MAC) → 网络层(IP) → 传输层(TCP/UDP/端口) → 应用层(HTTP/SMTP…).  
   • 核心概念：MAC 地址（硬件身份证）、IP 地址（全球定位）、端口（进程通道）、Socket（IP+端口）。  
   • 常见应用协议：DHCP、DNS、HTTP/HTTPS。  
   • 口诀：网线(物理)–网卡(链路)–路由(网络)–端口(传输)–应用程序(应用)。

4. 远程连接  
   • SSH：加密命令行登录，支持密钥认证、端口转发，配合 SCP/SFTP 传文件。  
   • VNC：跨平台图形桌面共享，可与 SSH 隧道组合保证安全。  
   • RDP、TeamViewer/Chrome Remote Desktop：其它图形远控方案。  
   • 典型场景：云服务器维护、IoT 设备管理、远程技术支持与协作。

5. 学习资源与协作平台  
   • 菜鸟教程 – 中文速查 & 在线跑代码。  
   • 中国大学 MOOC – 体系化课程，适合自学深造。  
   • Git / GitHub – 版本控制与全球开源协作；核心命令：init、add、commit、push、pull request。  
   • Gitee – 本土代码托管，网络更友好。  
   • Stack Overflow – 全球最大程序员问答库，“报错先搜 Stack”。  
   • LeetCode – 在线刷题与算法练习，面试/思维训练利器。
